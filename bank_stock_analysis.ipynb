```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
%matplotlib inline
sns.set_theme(color_codes=True)
import matplotlib.pyplot as plt
```

```python
# Set start and end date parameters
start = datetime(2006, 1, 1)
end = datetime(2016, 1, 1)
```

# Retrieving the Data

In this section we will be retrieving stock data from the following banks directly from the web using Pandas DataReader:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo

1. Retrieve info for Bank of America

```python
BAC_df = data.DataReader("BAC", 'yahoo', start, end)
BAC_df.head()
```

2. Retrieve info for CitiGroup

```python
C_df = data.DataReader("C", 'yahoo', start, end)
C_df.head()
```

3. Retrieve info for Goldman Sachs

```python
GS_df = data.DataReader("GS", 'yahoo', start, end)
GS_df.head()
```

4. Retrieve info for JPMorgan Chase

```python
JPM_df = data.DataReader("JPM", 'yahoo', start, end)
JPM_df.head()
```

5. Retrieve info for Morgan Stanley

```python
MS_df = data.DataReader("MS", 'yahoo', start, end)
MS_df.head()
```

6. Retrieve info for Wells Fargo

```python
WFC_df = data.DataReader("WFC", 'yahoo', start, end)
WFC_df.head()
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stocks = pd.concat([BAC_df, C_df, GS_df, JPM_df, MS_df, WFC_df], axis=1, keys=tickers)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# Exploring the Data

1. Calculating the max Close price for each bank's stock throughout the time period.

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

2. Creating a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

3. Use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

4. Create a pairplot using seaborn of the returns dataframe. What stock stands out to you? Can you figure out why?

```python
sns.pairplot(returns[1:])
```
