```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic_df = sns.load_dataset('titanic')
```

```python
titanic_df.head()
```

```python
# Create a Joint Plot based on fare and age
sns.jointplot(x='fare',y='age',data=titanic_df,kind='scatter')
```

```python
# Create a distribution plot on fare
sns.distplot(titanic_df['fare'],kde=False,bins=30, color="red")
```

```python
# Create a box plot based on class and age
sns.boxplot(x="class", y="age", data=titanic_df,palette='rainbow')
```

```python
# Create swarmplot based on class and age
sns.swarmplot(x="class", y="age", data=titanic_df)
```

```python
# Create a countplot of Males vs Females
sns.countplot(x='sex',data=titanic_df)
```

```python
titanic_df.corr()
```

```python
# Create a heatmap on the dataset
sns.heatmap(titanic_df.corr(),cmap='coolwarm',annot=False)
```

```python
# Create a facet grid on age and sex
g = sns.FacetGrid(data=titanic_df,col='sex')
g.map(plt.hist,'age')
```
